Выкинем из датасета данные, в какой-то мере дублирующие друг друга, оставим агрегирующие их признаки

```python
CBJ_data.drop(['SHG',
               'Power Play Opportunities',
               'Oponent SHG',
               'Opponent Power Play Opportunities',
               'CF = Shots + Blocks + Misses',
               'CA = Shots + Blocks + Misses Against',
               'FF = Shots + Misses',
               'FA = Shots + Misses Against',
               'Faceoff Wins',
               'Faceoff Losses',
               'Offensive Zone Faceoffs / (Offensive Zone Faceoffs + Defensive Zone Faceoffs)'],
              axis=1, inplace =True)
```

Видно, что есть пропуски по данным, их надо обработать.

```python
# 'Home' - можно представить как бинарный признак - была сыграда домашняя игра (1) или гостевая (0),
# заполним соответствующе, приведем к типу int
CBJ_data['Home'] = CBJ_data['Home'].fillna(0)
CBJ_data['Home'] = CBJ_data['Home'].astype('int16')

# 'Opponent Power Play Goals' - забитые голы противником при игре в большинстве,
#заполним пропуски 0, приведем к типу int
CBJ_data['Opponent Power Play Goals'] = CBJ_data['Opponent Power Play Goals'].fillna(0)
CBJ_data['Opponent Power Play Goals'] = CBJ_data['Opponent Power Play Goals'].astype('int16')
```

Колонку `Date` приведу к временному формату и сделаю индексы по датам.  
Для удобного агрегирования по сезонам добавляю колонку `Season`. Это будет категориальный признак.

```python
CBJ_data['Date'] = pd.to_datetime(CBJ_data.Date,format='%Y-%m-%d')
CBJ_data.index  = CBJ_data.Date

CBJ_data.loc[CBJ_data.Date['2014-8':'2015-7'], 'Season'] = 15
CBJ_data.loc[CBJ_data.Date['2015-8':'2016-7'], 'Season'] = 16
CBJ_data.loc[CBJ_data.Date['2016-8':'2017-7'], 'Season'] = 17
CBJ_data.loc[CBJ_data.Date['2017-8':'2018-7'], 'Season'] = 18
CBJ_data.loc[CBJ_data.Date['2018-8':'2019-7'], 'Season'] = 19
CBJ_data['Season'] = CBJ_data['Season'].astype('int16')
CBJ_data['Season'] = CBJ_data['Season'].astype('category')

CBJ_data.drop('Date', axis = 1, inplace = True)
```

Закодирую бинарный признак `Result` как 1 в случае победы (W) и 0 в случае поражения (L)

```python
d = {'W' : 1, 'L' : 0}
CBJ_data['Result'] = CBJ_data['Result'].map(d)
CBJ_data['Result'] = CBJ_data['Result'].astype('int16')
```

Данные по каждой игре имеют относительно небольшой размер, влезающий в int16 и float32. Приведу данные к этим типам.

```python
CBJ_data['Games Played'] = CBJ_data['Games Played'].astype('int16')
CBJ_data['Goals'] = CBJ_data['Goals'].astype('int16')
CBJ_data['Goals Against'] = CBJ_data['Goals Against'].astype('int16')
CBJ_data['Shots on Goal'] = CBJ_data['Shots on Goal'].astype('int16')
CBJ_data['Penalties in Minutes'] = CBJ_data['Penalties in Minutes'].astype('int16')
CBJ_data['Power Play Goals'] = CBJ_data['Power Play Goals'].astype('int16')
CBJ_data['Opponent Shots on Goal'] = CBJ_data['Opponent Shots on Goal'].astype('int16')
CBJ_data['Opponent Penalties in Minutes'] = CBJ_data['Opponent Penalties in Minutes'].astype('int16')

CBJ_data['CF / (CF + CA)'] = CBJ_data['CF / (CF + CA)'].astype('float32')
CBJ_data['FF / (FF + FA)'] = CBJ_data['FF / (FF + FA)'].astype('float32')
CBJ_data['Faceoff Win Percentage'] = CBJ_data['Faceoff Win Percentage'].astype('float32')
CBJ_data['Shooting % + Save %'] = CBJ_data['Shooting % + Save %'].astype('float32')
```

Название команды соперника буду использовать как категориальный признак

```python
CBJ_data['Opponent'] = CBJ_data['Opponent'].astype('category')
```

Проверим следующие гипотезы:

H0 - количеством забитых/пропущенных голов и номер сыгранного матча (период сезона: начало, середина, конец) не взаимосвязаны

H1 - есть зависимость между показателями

Проверяю данные на нормальность распределения используя притерий Шапиро-Уилка, для визуализации применяю Q-Q график:

```python
stats.shapiro(CBJ_data['Goals'])
```

```python
stats.probplot(CBJ_data['Goals'], dist="norm", plot=plt)
plt.show()
```

```python
stats.shapiro(CBJ_data['Goals Against'])
```

```python
stats.probplot(CBJ_data['Goals Against'], dist="norm", plot=plt)
plt.show()
```

Нормальности нет, будем использовать корреляцию Спирмена, коэффициент является методом непараметрического анализа

```python
stats.spearmanr(CBJ_data['Games Played'], CBJ_data['Goals'])
```

```python
stats.spearmanr(CBJ_data['Games Played'], CBJ_data['Goals Against'])
```

pvalue получился почти 13% (забитые голы) и 33% (пропущенные голы) соответственно, что говорит нам о подтверждении Н0 - параметры не коррелируют. correlation=0.0749 (забитые голы) и correlation=-0.0485 (пропущенные голы), учитывая, что коэф корреляции Спирмена находится в пределах [-1, 1], то близость полученного коэф. к 0 по нашим фичам подтверждает факт некоррелирования номера матча и количества забитых/пропущенных голов.  
Т.е мы не можем говорить о закономерности "выхода команды на игровую мощность в течении сезона".

Проверим значимо ли различаются выигранные и проигранные игры по количеству забитых сопернику шайб:

H0 - med(X1 - X2) == 0

H1 - med(X1 - X2) != 0

Сделаем выбори по проигранным и выигранным матчам, будем смотреть на кол-во забитых противнику голов

```python
goals_loss = CBJ_data.loc[CBJ_data['Result'] == 0, 'Goals']
goals_win = CBJ_data.loc[CBJ_data['Result'] == 1, 'Goals']

print(goals_loss.shape, goals_win.shape)
```

Использую непараметрический двувыборочный критерий знаков рангов Уилкоксона, поэтому есть ограничение на размер выборки до 50 экземпляров. Проверка на нормальность распределения не требуется.

```python
goals_win = goals_win[-50:]
goals_loss = goals_loss[-50:]

print(goals_loss.shape == goals_win.shape)
```

```python
stats.wilcoxon(goals_loss[-50:], goals_win[-50:])
```

Н0 не принимается, различия в выборках есть.

H1 - есть зависимость между показателями

Перед тем как применять статистическую модель для временного ряда необходимо убедиться, что ряд стационарен. Использую статистический тест.

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries, window = 7, cutoff = 0.05):

    #Determing rolling statistics
    rolmean = timeseries.rolling(window).mean()
    rolstd = timeseries.rolling(window).std()

    #Plot rolling statistics:
    fig = plt.figure(figsize=(10, 5))
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    
    plt.show()
    
    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries.values,autolag='AIC' )
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    pvalue = dftest[1]
    if pvalue < cutoff:
        print('p-value = %.4f. The series is likely stationary.' % pvalue)
    else:
        print('p-value = %.4f. The series is likely non-stationary.' % pvalue)
    
    print(dfoutput)
```

Частотность беру равную 82, так как результаты учитываются по сезонам, каждый из которых состоит из 82 игр

```python
test_stationarity(CBJ_data['Result'], window=82)
```

Ряд стационарен.

Чтобы построить модель нам нужно знать ее порядок, состоящий из 6-х параметров:

p,P — порядок компоненты AR

q,Q — порядок компонетны MA

```python
s = 82

ps = [3, 14, 25, 50, 64, 68, 72]
d=1
qs = [3, 6, 64, 70]
Ps = range(0, 1)
D=0
Qs = range(0, 1)
```

```python
from itertools import product

parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)
```

```python
%%time
import warnings
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')

for param in parameters_list:
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model=sm.tsa.statespace.SARIMAX(CBJ_data.Result, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], 82)).fit(disp=-1)
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
warnings.filterwarnings('default')
```

```python
stat_test = sm.tsa.adfuller(best_model.resid[:])
print ('adf: ', stat_test[0] )
print ('p-value: ', stat_test[1])
print('Critical values: ', stat_test[4])
if stat_test[0]> stat_test[4]['5%']: 
    print ('есть единичные корни, ряд не стационарен')
else:
    print ('единичных корней нет, ряд стационарен')
```
